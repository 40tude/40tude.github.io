```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# Recall values from 0 to 100%
recall_values = np.linspace(0.01, 1, 100)  # Start at 0.01 to avoid division by zero

# Fixed Precision values
precision_fixed = [0.20, 0.50, 0.90]
colors = ['red', 'blue', 'green']
labels = ['Precision = 20%', 'Precision = 50%', 'Precision = 90%']

# Create figure
plt.figure(figsize=(10, 6))

# Plot F1 score for each fixed Precision value
for i, prec in enumerate(precision_fixed):
    # Calculate F1 score: F1 = 2 * (Precision * Recall) / (Precision + Recall)
    f1_scores = 2 * (prec * recall_values) / (prec + recall_values)

    plt.plot(recall_values * 100, f1_scores,
             color=colors[i], linewidth=2, label=labels[i])

    # Mark the maximum point (when Recall = Precision)
    max_f1 = 2 * (prec * prec) / (prec + prec)
    plt.plot(prec * 100, max_f1, 'o', color=colors[i],
             markersize=10, markeredgecolor='black', markeredgewidth=1.5)
    plt.annotate(f'Max: ({prec*100:.0f}%, {max_f1:.2f})',
                xy=(prec * 100, max_f1),
                xytext=(10, 10), textcoords='offset points',
                fontsize=9, color=colors[i], fontweight='bold')

plt.xlabel('Recall (%)', fontsize=12)
plt.ylabel('F1 Score', fontsize=12)
plt.title('F1 Score vs Recall for Different Fixed Precision Values\n(Maximum occurs when Recall = Precision)',
          fontsize=13, fontweight='bold')
plt.grid(True, alpha=0.3, linestyle='--')
plt.legend(fontsize=11, loc='lower right')
plt.xlim(0, 100)
plt.ylim(0, 1)
plt.tight_layout()
plt.show()

print("Key observation:")
print("=" * 60)
for prec in precision_fixed:
    max_f1 = prec  # When Recall = Precision, F1 = that value
    print(f"Precision = {prec*100:>3.0f}%  Max F1 = {max_f1:.2f} at Recall = {prec*100:.0f}%")
```
